```python
def metrics(y_test, predict):
    acc = accuracy_score(y_test, predict)
    precision = precision_score(y_test, predict)
    recall = recall_score(y_test, predict)
    f1 = f1_score(y_test, predict)
    roc_score = roc_auc_score(y_test, predict)
    
    print('정확도 : {0:.2f}, 정밀도 : {1:.2f}, 재현율 : {2:.2f}'.format(acc, precision, recall))
    print('f1-score : {0:.2f}, auc : {1:.2f}'.format(f1, roc_score))
```

```python
scailing_df = origin_df.drop(labels='Amount', axis=1)
scailing_df['Amoumt'] = np.log1p(origin_df['Amount'])
scailing_df
```

```python
X = origin_df.iloc[:, :-1]
y = origin_df.iloc[:, -1]

X_al = scailing_df[[col for col in scailing_df.columns if col != 'Class']]
y_al = scailing_df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=0, stratify = y)
X_al_train, X_al_test, y_al_train, y_al_test = train_test_split(X_al, y_al, test_size = 0.25, random_state=0, stratify = y)
```

```python
print(y_train.value_counts() / y_train.count())
print(y_test.value_counts() / y_test.count())

print(y_al_train.value_counts() / y_al_train.count())
print(y_al_test.value_counts() / y_al_test.count())
```

```python
def modeling(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    metrics(y_test, pred)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
modeling(lr, X_train, X_test, y_train, y_test)
```

```python
lr = LogisticRegression()
modeling(lr, X_al_train, X_al_test, y_al_train, y_al_test)
```

```python
lgb = LGBMClassifier(n_estimators = 1000, num_leaves = 64, n_jobs = -1, boost_from_average = False)
modeling(lgb, x_train = X_train, x_test = X_test, y_train = y_train, y_test = y_test)
```

```python
lgb = LGBMClassifier(n_estimators = 1000, num_leaves = 64, n_jobs = -1, boost_from_average = False)
modeling(lgb, x_train = X_al_train, x_test = X_al_test, y_train = y_al_train, y_test = y_al_test)
```
